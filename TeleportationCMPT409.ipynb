```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
circuit = QuantumCircuit (3,3)
```

```python
circuit.x(0) #Qubit 0 is the qubit to be teleported.
circuit.barrier()
```

```python

circuit.h(1)  #Create EPR Pair/bellstate
circuit.cx(1,2)
```

```python
circuit.cx(0,1)   #Teleporting
circuit.h(0)
circuit.barrier()
```

```python
circuit.measure([0,1],[0,1])    #Collapse and measure the qubits 0 and 1
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.cx(1,2)  #Apply CNOT Gate
circuit.cz(0,2)  #Apply Pauli Z Gate
```

```python

circuit.measure(2,2)  #Collapse and mesaure the qubit that is recieivng the teleported qubit( qubit 0)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots=1000).result()
counts = job.get_counts()
plot_histogram(counts)
```
